```python
크롤링(Crawling) : web 상에 존재하는 contents를 수집하는 작업
    - HTML 페이지를 가져와서 HTML/CSS 등을 파싱하고 필요한 데이터만 추출하는 기법
    - Open API(REST API)를 제공하는 서비스에 Open API를 호출해서 받은 데이터 중 필요한 데이터만 추출하는 기법
    - selenium 등 브라우저를 프로그래밍으로 조작해서 필요한 데이터만 추출하는 기법
```

```python
BeautifulSoul 라이브러리 활용
   - HTML의 태그를 파싱해서 필요한 데이터만 추출하는 함수를 제공하는 라이브러리
   - pip install bs4
```

```python
import requests
from bs4 import BeautifulSoup

# 1) requests 라이브러리를 활용한 HTML 페이지 요청
# 2) req 객체에 HTML 데이터가 저장되고, req.content 로 데이터를 추출할 수 있음

# naver.com 접속해서 enter 치기
req = requests.get('http://www.naver.com')

# HTML 에 있는 코드 가져와서 parsing
soup = BeautifulSoup(req.text, "html.parser")

#3) HTML 페이지 파싱 - BeautifulSoup(HTML데이터, 파싱방법)
soup = BeautifulSoup(req.text, "html.parser")

#4) 필요한 데이터 검색 - ex
title = soup.find('title')

#5) 필요한 데이터 추출
print(title.get_text())
```

```python
find() / find_all()
find() : 가장 먼저 검색되는 태그 반환
find_all() : 검색되는 전체 태그 반환
```

## 다음 웹 크롤링

```python
import requests
from bs4 import BeautifulSoup

# naver.com 접속해서 enter 치기
req = requests.get('http://www.daum.net')

# HTML 에 있는 코드 가져와서 parsing
soup = BeautifulSoup(req.text, "html.parser")
# 실시간 이슈 검색어만 추출
#print(soup)
# parsing 하고 특정태그의 텍스트 부분만 출력
# find_all => tag 와 class_= 을 활용find_all("span", class_="link_txt")
#  select ==> css_selector 이용
lines = soup.select('.link_txt')
for line in lines:
    print(line.get_text())
```

## 네이트 판 크롤링

```python
import requests
from bs4 import BeautifulSoup

# naver.com 접속해서 enter 치기
req = requests.get('https://pann.nate.com/talk')

# HTML 에 있는 코드 가져와서 parsing
soup = BeautifulSoup(req.text, "html.parser")
# 실시간 이슈 검색어만 추출
#print(soup)

# parsing 하고 특정태그의 텍스트 부분만 출력
# find_all => tag 와 class_= 을 활용find_all("span", class_="link_txt")
#  select ==> css_selector 이용
lines = soup.select('#todayTalk_0 > li')
for line in lines:
    print(line.find("a").text)
```
